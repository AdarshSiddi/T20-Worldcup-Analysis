```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("/kaggle/input/t20-worldcup-analysis20222/T20 WORLDCUP STATS ANALYSIS-2022.csv")
data 
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='TOSS WINNER',data=data)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='WINNER',data=data)
```

```python


WON_BY = data["WON BY"].value_counts()
label = WON_BY.index
counts = WON_BY.values
colors = ['ORANGE','GREEN']
plt.figure(figsize=(5,10))

plt.pie(counts,labels=label,colors=colors,autopct='%1i%%',shadow=True)
plt.legend(title='Number of Matches Won By Runs Or Wickets')
plt.axis('equal')
plt.show()
```

```python

TOSS = data["TOSS DECISION"].value_counts()
label = TOSS.index
counts = TOSS.values
colors = ['skyblue','lightgreen']
plt.figure(figsize=(6,10))

plt.pie(counts,labels=label,colors=colors,autopct='%1.0f%%',shadow=True)
plt.legend(title='Toss Decisions in this Worldcup')
plt.axis('equal')
plt.show()

```
